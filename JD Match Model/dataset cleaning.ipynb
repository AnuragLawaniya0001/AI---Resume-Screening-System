```python
import pandas
```

```python
df = pandas.read_csv("resume_data.csv",encoding='utf-8-sig')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(["address","educational_results","result_types", "company_urls","extra_curricular_organization_links", "online_links"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(["issue_dates"	,"expiry_dates"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df["passing_years"] = df["passing_years"].apply(lambda x: f"Passing Year: {x}")
df["degree_names"] = df["degree_names"].apply(lambda x: f"degree names: {x}")
df["age_requirement"] = df["age_requirement"].apply(lambda x: f"age requirement: {x}")
df["responsibilities.1"] = df["responsibilities.1"].apply(lambda x: f"responsibilities : {x}")
```

```python
df.head()
```

```python
resume_cols = [
    'career_objective', 'skills', 'educational_institution_name',
    'degree_names', 'passing_years', 'major_field_of_studies',
    'professional_company_names', 'start_dates', 'end_dates',
    'related_skils_in_job', 'positions', 'locations', 'responsibilities',
    'extra_curricular_activity_types', 'extra_curricular_organization_names',
    'role_positions', 'languages', 'proficiency_levels',
    'certification_providers', 'certification_skills'
]

jd_cols = [
    '﻿job_position_name', 'educationaL_requirements',
    'experiencere_requirement', 'age_requirement', 
    'responsibilities.1', 'skills_required'
]
# Join columns while explicitly forcing everything to be a string and ignoring NaNs
df['resume_text'] = df[resume_cols].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
df['jd_text'] = df[jd_cols].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

```python
print(df[['resume_text', 'jd_text']].head())
```

```python


junk_chars = r'[\[\]"\',\n\r\t]'

df['resume_text'] = df['resume_text'].str.replace(junk_chars, ' ', regex=True)
df['jd_text'] = df['jd_text'].str.replace(junk_chars, ' ', regex=True)
```

```python
df['resume_text'] = df['resume_text'].str.lower()
df['jd_text'] = df['jd_text'].str.lower()
```

```python
junk_words = r'\b(n/a|nan|none)\b'
df['resume_text'] = df['resume_text'].str.replace(junk_words, '', regex=True)
df['jd_text'] = df['jd_text'].str.replace(junk_words, '', regex=True)
```

```python
df['resume_text'] = df['resume_text'].str.replace(r'\s+', ' ', regex=True).str.strip()
df['jd_text'] = df['jd_text'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
print(df[['resume_text', 'jd_text']].head())
```

```python
keep_cols = ['resume_text', 'jd_text', 'matched_score']
df = df[keep_cols]
```

```python
print(df.info())
print(df.head())
```

```python
df.to_csv('cleaned_resume_jd_dataset_2', index=False)
```
